```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Matrix size
height, width = 100, 100

# Create a random grayscale matrix
matrix = np.random.randint(0, 256, (height, width), dtype=np.uint8)
```

```python
# Convert matrix to grayscale image
img = Image.fromarray(matrix, mode='L')

# Display using Pillow
img.show()  # Will open image in Colab's viewer

# Display inline with matplotlib
plt.imshow(matrix, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img.save('grayscale_output1.png')
print("Image saved as grayscale_output.png")
```

Creating simple patterns


```python
matrix = np.zeros((height, width), dtype=np.uint8)
for i in range(height):
    for j in range(width):
        matrix[i, j] = int((j / width) * 255)  # gradient left to right

plt.imshow(matrix, cmap='gray')
plt.axis('off')
plt.show()
```

checkboard pattern

```python
matrix = np.zeros((height, width), dtype=np.uint8)
block_size = 10
for i in range(height):
    for j in range(width):
        if (i//block_size + j//block_size) % 2 == 0:
            matrix[i, j] = 0
        else:
            matrix[i, j] = 255

plt.imshow(matrix, cmap='gray')
plt.axis('off')
plt.show()
```
